```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
db1 = pd.read_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB1/wdbc.csv')
db1
```

```python
db1.describe()
```

```python
# Extract the second column
second_column = db1.pop(db1.columns[1])

# Add the second column to the end of the DataFrame
db1.insert(len(db1.columns), second_column.name, second_column)
db1
```

```python
# Extract column names from the DataFrame
column_names = db1.columns.tolist()

# Create a new DataFrame with the duplicated column names as the first row
new_row = pd.DataFrame([column_names], columns=db1.columns)

# Concatenate the new row and the rest of the data
db1 = pd.concat([new_row, db1], ignore_index=True)
# Rename the column names
new_column_names = [f"X.{i}" for i in range(len(column_names) - 1)] + ["Class"] 
db1.columns = new_column_names
# Convert all columns except "Class" to numeric data types
for column in db1.columns:
    if column != "Class":
        db1[column] = pd.to_numeric(db1[column], errors="coerce")
db1
```

```python
db1.hist(bins=20, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
db1.describe()
```

```python
from sklearn.preprocessing import StandardScaler
# Separate the target variable ('Class') from the features
target = db1['Class']
features = db1.drop(columns=['Class'])

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the features and transform them
scaled_features = scaler.fit_transform(features)

# Create a DataFrame with the scaled features and original index
std_db1 = pd.DataFrame(scaled_features, columns=features.columns, index=features.index)

# Concatenate the target variable and scaled features back together
std_db1 = pd.concat([target, std_db1], axis=1)
std_db1
```

```python
std_db1.to_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB1/std_db1.csv', index=False)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Extract the Class column
class_column = db1.pop('Class')

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numeric columns
norm_db1 = pd.DataFrame(scaler.fit_transform(db1), columns=db1.columns)

# Add back the Class column
norm_db1 = pd.concat([class_column, norm_db1], axis=1)
norm_db1
```

```python
norm_db1.to_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB1/norm_db1.csv', index=False)
```
