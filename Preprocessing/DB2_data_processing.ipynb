```python
import pandas as pd
import matplotlib.pyplot as plt
db2 = pd.read_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB2/breast-cancer.csv')
db2
```

```python
# Extract the second column
column = db2.pop(db2.columns[0])

# Add the second column to the end of the DataFrame
db2.insert(len(db2.columns), column.name, column)
# Rename the 'no.1' column to 'no'
db2.rename(columns={'no.1': 'no'}, inplace=True)
db2
```

```python
# Extract column names from the DataFrame
column_names = db2.columns.tolist()

# Create a new DataFrame with the duplicated column names as the first row
new_row = pd.DataFrame([column_names], columns=db2.columns)

# Concatenate the new row and the rest of the data
db2 = pd.concat([new_row, db2], ignore_index=True)
# Rename the column names
new_column_names = ['age']+['menopause']+['tumor_size']+['inv_nodes']+['node_caps']+['deg_malig']+['breast']+['breast_quad']+['irradiat']+ ["Class"] 
db2.columns = new_column_names
# Convert 'deg_malig' column to numeric data type
db2['deg_malig'] = pd.to_numeric(db2['deg_malig'], errors='coerce')
db2
```

```python
db2.isna().sum()
```

```python
# Display unique values of all columns
for column in db2.columns:
    unique_values = db2[column].unique()
    print(f"Unique values in column '{column}': {unique_values}")
```

```python
# Display count of "?" in each column
for column in db2.columns:
    question_count = (db2 == "?").sum()
question_count
```

```python
db2.mode().iloc[0]
```

```python
# Display frequency table for each column
for column in db2.columns:
    frequency_table = db2[column].value_counts()
    print(f"Frequency table for column '{column}':")
    print(frequency_table)
    print("\n")
```

```python
db2['node_caps'].replace("?", "no", inplace=True)
db2['breast_quad'].replace("?", "left_low", inplace=True)
for column in db2.columns:
    unique_values = db2[column].unique()
    print(f"Unique values in column '{column}': {unique_values}")
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Iterate through categorical columns and apply label encoding
categorical_columns = ['age', 'menopause', 'tumor_size', 'inv_nodes', 'node_caps', 'breast', 'breast_quad', 'irradiat','Class']
for col in categorical_columns:
    db2[col] = label_encoder.fit_transform(db2[col])
db2
```

```python
db2.info()
```

```python
from sklearn.preprocessing import StandardScaler
# Extract the features from the DataFrame
features = db2.drop(columns=['Class'])

# Initialize the StandardScaler
scaler = StandardScaler()

# Perform data standardization on the features
features_standardized = scaler.fit_transform(features)

# Convert the standardized features back to a DataFrame
features_standardized_df = pd.DataFrame(features_standardized, columns=features.columns)

# Combine the standardized features with the 'Class' column
db2_standardized = pd.concat([db2['Class'], features_standardized_df], axis=1)
db2_standardized
```

```python
from sklearn.preprocessing import MinMaxScaler

# Extract the Class column
class_column = db2.pop('Class')

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numeric columns
norm_db2 = pd.DataFrame(scaler.fit_transform(db2), columns=db2.columns)

# Add back the Class column
norm_db2 = pd.concat([class_column, norm_db2], axis=1)
norm_db2
```
