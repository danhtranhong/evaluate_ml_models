```python
import pandas as pd
import matplotlib.pyplot as plt
db3 = pd.read_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB3/mammographic_masses.csv')
db3
```

```python
db3_txt = pd.read_table(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB3/mammographic_masses.csv')
db3_txt
```

```python
# Extract column names from the DataFrame
column_names = db3.columns.tolist()

# Create a new DataFrame with the duplicated column names as the first row
new_row = pd.DataFrame([column_names], columns=db3.columns)

# Concatenate the new row and the rest of the data
db3 = pd.concat([new_row, db3], ignore_index=True)
# Rename the column names
new_column_names = [f"X.{i}" for i in range(len(column_names) - 1)] + ["Class"] 
db3.columns = new_column_names
# Convert all columns except "Class" to numeric data types
for column in db3.columns:
    if column != "Class":
        db3[column] = pd.to_numeric(db3[column], errors="coerce")
db3
```

```python
# Change specific values in the DataFrame
db3.at[0, 'X.3'] = 5.0
db3.at[0, 'X.4'] = 3.0
db3
```

```python
db3.isna().sum()
```

```python
#remove missing values
db3mod = db3.dropna()
# Reset the index of the cleaned DataFrame
db3mod = db3mod.reset_index(drop=True)
db3mod
```

```python
db3mod.info()
```

```python
db3mod['Class'] = pd.to_numeric(db3mod['Class'], errors="coerce")
db3mod.info()
```

```python
db3mod.describe()
```

```python
db3mod.to_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB3/DB3MOD.csv', index = False)
```

```python
from sklearn.preprocessing import StandardScaler
# Extract the features from the DataFrame
features = db3mod.drop(columns=['Class'])

# Initialize the StandardScaler
scaler = StandardScaler()

# Perform data standardization on the features
features_standardized = scaler.fit_transform(features)

# Convert the standardized features back to a DataFrame
features_standardized_df = pd.DataFrame(features_standardized, columns=features.columns)

# Combine the standardized features with the 'Class' column
db3_standardized = pd.concat([db3mod['Class'], features_standardized_df], axis=1)
db3_standardized
```

```python
db3_standardized.to_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB3/std_db3.csv', index=False)
```

```python
db3_standardized.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Extract the Class column
class_column = db3mod.pop('Class')

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numeric columns
norm_db3 = pd.DataFrame(scaler.fit_transform(db3mod), columns=db3mod.columns)

# Add back the Class column
norm_db3 = pd.concat([class_column, norm_db3], axis=1)
norm_db3
```
