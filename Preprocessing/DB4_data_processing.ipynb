```python
import pandas as pd
import matplotlib.pyplot as plt
db4 = pd.read_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB4/breast-cancer-wisconsin.csv')
db4
```

```python
db4_txt = pd.read_table(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB4/breast-cancer-wisconsin.csv')
db4_txt
```

```python
# Extract column names from the DataFrame
column_names = db4.columns.tolist()

# Create a new DataFrame with the duplicated column names as the first row
new_row = pd.DataFrame([column_names], columns=db4.columns)

# Concatenate the new row and the rest of the data
db4 = pd.concat([new_row, db4], ignore_index=True)
# Rename the column names
new_column_names = [f"X.{i}" for i in range(len(column_names) - 1)] + ["Class"] 
db4.columns = new_column_names
# Convert all columns except "Class" to numeric data types
for column in db4.columns:
    if column != "Class":
        db4[column] = pd.to_numeric(db4[column], errors="coerce")
db4
```

```python
# Change specific values in the DataFrame
db4.at[0, 'X.3'] = 1.0
db4.at[0, 'X.4'] = 1.0
db4.at[0, 'X.6'] = 1.0
db4.at[0, 'X.8'] = 1.0
db4.at[0, 'X.9'] = 1.0
db4.at[0, 'Class'] = 2
db4
```

```python
db4.isna().sum()
```

```python
# Display unique values of all columns
for column in db4.columns:
    unique_values = db4[column].unique()
    print(f"Unique values in column '{column}': {unique_values}")
```

```python
#removing missing value
db4mod = db4.dropna()
# Reset the index of the cleaned DataFrame
db4mod = db4mod.reset_index(drop=True)
db4mod
```

```python
db4mod.info()
```

```python
db4mod['Class'] = pd.to_numeric(db4mod['Class'], errors="coerce")
db4mod.info()
```

```python
db4mod.describe()
```

```python
db4mod.to_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB4/DB4MOD.csv', index = False)
```

```python
from sklearn.preprocessing import StandardScaler
# Extract the features from the DataFrame
features = db4mod.drop(columns=['Class'])

# Initialize the StandardScaler
scaler = StandardScaler()

# Perform data standardization on the features
features_standardized = scaler.fit_transform(features)

# Convert the standardized features back to a DataFrame
features_standardized_df = pd.DataFrame(features_standardized, columns=features.columns)

# Combine the standardized features with the 'Class' column
db4_standardized = pd.concat([db4mod['Class'], features_standardized_df], axis=1)
db4_standardized
```

```python
db4_standardized.describe()
```

```python
db4_standardized.to_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB4/std_db4.csv', index=False)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Extract the Class column
class_column = db4mod.pop('Class')

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numeric columns
norm_db4 = pd.DataFrame(scaler.fit_transform(db4mod), columns=db4mod.columns)

# Add back the Class column
norm_db4 = pd.concat([class_column, norm_db4], axis=1)
norm_db4
```
