```python
import pandas as pd

# Open the GDS360_full.soft file in read mode
with open('/Users/minhnguyetnguyen/Downloads/GDS360_full.soft', 'r') as file:
    # Read the lines of the file
    lines = file.readlines()

# Process the lines and extract data
data = []
for line in lines:
    # Split the line based on the tab character
    values = line.strip().split('\t')
    data.append(values)

# Create a DataFrame from the extracted data
df = pd.DataFrame(data)

# Save the DataFrame as a CSV file
df.to_csv('GDS360_full.csv', index=False)

df
```

```python
df.head(86)
```

```python
# Delete the first 86 rows
df = df.iloc[86:]

# Reset the index
df.reset_index(drop=True, inplace=True)
df
```

```python
# Delete columns from the 27th column to the end
df = df.iloc[:, :26]
df = df.drop(df.columns[0], axis=1)
df
```

```python
# Set the first row as the column names
df.columns = df.iloc[0]

# Drop the first row (since it's now the column names)
df = df[1:]
df
```

```python
# Sort by the 'IDENTIFIER' column in ascending order
df = df.sort_values(by='IDENTIFIER')
df = df.dropna()
df
```

```python
numeric_cols = df.columns.difference(['IDENTIFIER'])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df.info()
```

```python
df5 = df.groupby('IDENTIFIER', as_index=False).mean()
df5
```

```python
# Set the "IDENTIFIER" column as the index
df5.set_index("IDENTIFIER", inplace=True)
df5
```

```python
db5 = df5.transpose()
db5
```

```python
db5['Disease_state'] = db5.index
db5.reset_index(drop=True, inplace=True)
db5
```

```python
# Replace values in the "disease_state" column
db5['Disease_state'].replace({
    'GSM4901':'docetaxel resistant tumor',
    'GSM4902':'docetaxel resistant tumor',
    'GSM4904':'docetaxel resistant tumor',
    'GSM4905':'docetaxel resistant tumor',
    'GSM4906':'docetaxel resistant tumor',
    'GSM4909':'docetaxel resistant tumor',
    'GSM4910':'docetaxel resistant tumor',
    'GSM4911':'docetaxel resistant tumor',
    'GSM4912':'docetaxel resistant tumor',
    'GSM4913':'docetaxel resistant tumor',
    'GSM4916':'docetaxel resistant tumor',
    'GSM4918':'docetaxel resistant tumor',
    'GSM4922':'docetaxel resistant tumor',
    'GSM4924':'docetaxel resistant tumor',
    
    'GSM4903':'docetaxel sensitive tumor',
    'GSM4907':'docetaxel sensitive tumor',
    'GSM4908':'docetaxel sensitive tumor',
    'GSM4914':'docetaxel sensitive tumor',
    'GSM4915':'docetaxel sensitive tumor',
    'GSM4917':'docetaxel sensitive tumor',
    'GSM4919':'docetaxel sensitive tumor',
    'GSM4920':'docetaxel sensitive tumor',
    'GSM4921':'docetaxel sensitive tumor',
    'GSM4923':'docetaxel sensitive tumor'
}, inplace=True)
db5
```

```python
from sklearn.preprocessing import MinMaxScaler

# Extract the Class column
class_column = db5.pop('Disease_state')

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numeric columns
norm_db5 = pd.DataFrame(scaler.fit_transform(db5), columns=db5.columns)

# Add back the Class column
norm_db5 = pd.concat([class_column, norm_db5], axis=1)
norm_db5
```

```python
norm_db5.describe()
```

```python
from sklearn.decomposition import PCA

# Create a PCA instance with 8 components
pca = PCA(n_components=8)

# Extract the 'Disease_state' column for later
disease_state = norm_db5['Disease_state']

# Perform PCA on the numeric columns
numeric_data = norm_db5.drop(columns=['Disease_state'])  # Remove 'Disease_state' for PCA
pca_result = pca.fit_transform(numeric_data)

# Create a new DataFrame with the PCA results and the 'Disease_state' column
pca_db5 = pd.DataFrame(data=pca_result, columns=[f'PCA_{i+1}' for i in range(8)])
pca_db5['Disease_state'] = disease_state
pca_db5
```
