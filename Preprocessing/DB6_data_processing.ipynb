```python
# Get the column names as a list
columns = db6.columns.tolist()

# Move the second column to the end
second_column = columns.pop(1)
columns.append(second_column)

# Reorder the columns in the DataFrame
db6 = db6[columns]
db6
```

```python
# Extract column names from the DataFrame
column_names = db6.columns.tolist()

# Create a new DataFrame with the duplicated column names as the first row
new_row = pd.DataFrame([column_names], columns=db6.columns)

# Concatenate the new row and the rest of the data
db6 = pd.concat([new_row, db6], ignore_index=True)
# Rename the column names
new_column_names = [f"X.{i}" for i in range(len(column_names) - 1)] + ["Class"] 
db6.columns = new_column_names
# Convert all columns except "Class" to numeric data types
for column in db6.columns:
    if column != "Class":
        db6[column] = pd.to_numeric(db6[column], errors="coerce")
db6
```

```python
# Change specific values in the DataFrame
db6.at[0, 'X.33'] = 5.0
db6
```

```python
db6.info()
```

```python
db6.isna().sum()
```

```python
db6 = db6.dropna()
# Reset the index
db6 = db6.reset_index(drop=True)
db6
```

```python
# List of column names to check (exclude the "Class" column)
columns_to_check = [col for col in db6.columns if col != 'Class']

# Use applymap and str.isdigit to check if data in each cell is a digit
is_digit_df = db6[columns_to_check].applymap(lambda x: str(x).isdigit())

# Check if all columns have all digits (True) or not (False)
are_all_digits = is_digit_df.all()

# Display which columns contain non-digits
non_digit_columns = are_all_digits[~are_all_digits].index.tolist()
non_digit_columns
```

```python
from sklearn.preprocessing import StandardScaler
# Extract the features from the DataFrame
features = db6.drop(columns=['Class'])

# Initialize the StandardScaler
scaler = StandardScaler()

# Perform data standardization on the features
features_standardized = scaler.fit_transform(features)

# Convert the standardized features back to a DataFrame
features_standardized_df = pd.DataFrame(features_standardized, columns=features.columns)

# Combine the standardized features with the 'Class' column
db6_standardized = pd.concat([db6['Class'], features_standardized_df], axis=1)
db6_standardized
```

```python
from sklearn.preprocessing import MinMaxScaler

# Extract the Class column
class_column = db6.pop('Class')

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numeric columns
norm_db6 = pd.DataFrame(scaler.fit_transform(db6), columns=db6.columns)

# Add back the Class column
norm_db6 = pd.concat([class_column, norm_db6], axis=1)
norm_db6
```
