```python
import pandas as pd
import matplotlib.pyplot as plt
db7 = pd.read_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB7/DB7.csv')
db7
```

```python
# Select specific columns
db7mod = db7[['Age', 'Tumor Size', 'Reginol Node Positive', 'Survival Months', 'Regional Node Examined', 'Status']]
db7mod
```

```python
# List of column names to check (exclude the "Class" column)
columns_to_check = [col for col in db7mod.columns if col != 'Status']

# Use applymap and str.isdigit to check if data in each cell is a digit
is_digit_df = db7mod[columns_to_check].applymap(lambda x: str(x).isdigit())

# Check if all columns have all digits (True) or not (False)
are_all_digits = is_digit_df.all()

# Display which columns contain non-digits
non_digit_columns = are_all_digits[~are_all_digits].index.tolist()
non_digit_columns
```

```python
from sklearn.preprocessing import StandardScaler
# Extract the features from the DataFrame
features = db7mod.drop(columns=['Status'])

# Initialize the StandardScaler
scaler = StandardScaler()

# Perform data standardization on the features
features_standardized = scaler.fit_transform(features)

# Convert the standardized features back to a DataFrame
features_standardized_df = pd.DataFrame(features_standardized, columns=features.columns)

# Combine the standardized features with the 'Class' column
db7_standardized = pd.concat([db7mod['Status'], features_standardized_df], axis=1)
db7_standardized
```

```python
db7_standardized.to_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB7/std_db7.csv', index=False)
```

```python
db7_standardized.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Extract the Class column
class_column = db7mod.pop('Status')

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numeric columns
norm_db7 = pd.DataFrame(scaler.fit_transform(db7mod), columns=db7mod.columns)

# Add back the Class column
norm_db7 = pd.concat([class_column, norm_db7], axis=1)
norm_db7
```
