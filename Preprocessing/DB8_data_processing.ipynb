```python
import pandas as pd
import matplotlib.pyplot as plt
db8 = pd.read_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB8/dataR2.csv')
db8
```

```python
db8 = db8.rename(columns={'Classification': 'Class'})
db8
```

```python
from sklearn.preprocessing import StandardScaler
# Extract the features from the DataFrame
features = db8.drop(columns=['Class'])

# Initialize the StandardScaler
scaler = StandardScaler()

# Perform data standardization on the features
features_standardized = scaler.fit_transform(features)

# Convert the standardized features back to a DataFrame
features_standardized_df = pd.DataFrame(features_standardized, columns=features.columns)

# Combine the standardized features with the 'Class' column
db8_standardized = pd.concat([db8['Class'], features_standardized_df], axis=1)
db8_standardized
```

```python
db8_standardized.to_csv(r'/Users/minhnguyetnguyen/Documents/AISOUP/AutoML/Database/DB8/std_db8.csv', index=False)
```

```python
db8_standardized.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Extract the Class column
class_column = db8.pop('Class')

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numeric columns
norm_db8 = pd.DataFrame(scaler.fit_transform(db8), columns=db8.columns)

# Add back the Class column
norm_db8 = pd.concat([class_column, norm_db8], axis=1)
norm_db8
```

```python
norm_db8.describe()
```
